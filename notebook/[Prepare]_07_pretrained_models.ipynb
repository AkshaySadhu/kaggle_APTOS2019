# Pretrain Models for pytorch

```python
import torch
from efficientnet_pytorch import EfficientNet

PRE_TRAIN_MODEL_DIR = "../model/pretrain/"
```

```python
torch.hub.list("pytorch/vision", force_reload=True)
```

### DenseNet

```python
model = torch.hub.load('pytorch/vision', 'densenet121', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "densenet121.pth")
print(model)
```

### Inception V3

```python
model = torch.hub.load('pytorch/vision', 'inception_v3', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "inception_v3.pth")
print(model)
```

### MobileNet V2

```python
model = torch.hub.load('pytorch/vision', 'mobilenet_v2', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "mobilenet_v2.pth")
print(model)
```

### ResNet

```python
model = torch.hub.load('pytorch/vision', 'resnet18', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "resnet18.pth")
print(model)
```

```python
model = torch.hub.load('pytorch/vision', 'resnet34', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "resnet34.pth")
print(model)
```

```python
model = torch.hub.load('pytorch/vision', 'resnet50', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "resnet50.pth")
print(model)
```

```python
model = torch.hub.load('pytorch/vision', 'resnet101', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "resnet101.pth")
print(model)
```

### EfficientNet

ref : https://github.com/lukemelas/EfficientNet-PyTorch

```python
model = EfficientNet.from_pretrained("efficientnet-b5")
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "efficientnet-b5.pth")
print(model)
```
