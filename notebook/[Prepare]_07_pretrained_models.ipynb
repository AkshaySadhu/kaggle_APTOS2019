# Pretrain Models for pytorch

```python
import torch

PRE_TRAIN_MODEL_DIR = "../model/pretrain/"
```

```python
torch.hub.list("pytorch/vision", force_reload=True)
```

### MobileNet V2

```python
model = torch.hub.load('pytorch/vision', 'mobilenet_v2', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "mobilenet_v2.pth")
print(model)
```

### ResNet

```python
model = torch.hub.load('pytorch/vision', 'resnet18', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "resnet18.pth")
print(model)
```

```python
model = torch.hub.load('pytorch/vision', 'resnet34', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "resnet34.pth")
print(model)
```

```python
model = torch.hub.load('pytorch/vision', 'resnet50', pretrained=True)
torch.save(model.state_dict(), PRE_TRAIN_MODEL_DIR + "resnet50.pth")
print(model)
```
